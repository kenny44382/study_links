#### Using Terminal
- Split by 1000 lines

    + for i in *; do mv "$i" "$i.csv"; done
- Split by Bytes
    + split -b 50K bigdata.csv
- Split by Numeric/Suffix/Digit
    + split -d bigdata.csv
- Split and use custom name
    + split -l 1000 bigdata.csv mysmalldata_
- Split with N chunks
    + split -n5 bigdata.csv
