```python
# Load Pkgs
import pandas as pd
```

```python
# Load Our dataset
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
# Check For Missing NAN
df.isnull()
```

```python
# Sum of Missing NANA
df.isnull().sum()
```

#### Using Badfish
+ Counts
+ Pattern

```python
# Load Pkg
import badfish as bf
```

```python
dir(bf)
```

```python
miss_df = bf.MissFrame(df)
```

```python
# Count the Number of Missing NAN In Entire dataset
# Similar to df.isnull().sum()
miss_df.counts()
```

```python
# Count Number of Missing NAN in Columns
miss_df.counts(columns=['Salary','Age'])
```

```python
# Show Pattern
# Similar to df.isnull()
miss_df.pattern()
```

### Visualize The Pattern of NAN
+ Red/True : Missing
+ Blue/False : Not Missing

```python
# Visualize The Pattern of NAN
miss_df.plot(kind='pattern')
```

```python
# Using Where with Plot
miss_df.plot(kind='pattern',where=['Gender', 'Salary'])
```

### Using Dora
+ for data cleaning, 
+ feature selection & extraction, 
+ visualization, partitioning data for model validation, and 
+ versioning transformations of data.

```python
# Load the Pkgs
from Dora import Dora
```

```python
# Load and Initialize
dora = Dora(data = df)
```

```python
dir(dora)
```

```python
# To View Or Explore Data
dora.data
```

```python
dora.data['Age']
```

```python
# Drop A Column/Feature
dora.remove_feature('Address')
```

```python
dora.data.head()
```

```python
# Check For Missing NAN
dora.data.isnull()
```

```python
# Type is DataFrame Hence You Can Use it as such
type(dora.data)
```

```python
# To Use Functions You Need to Set Output
dora2 = Dora(output=0,data = df)
```

```python
dora2.data.head()
```
