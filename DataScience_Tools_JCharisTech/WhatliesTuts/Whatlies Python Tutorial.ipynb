### Word Embeddings
+ A word embedding is a learned representation for text where words that have the same meaning have a similar representation.
+ Word Embedding: refers to numerical representations of text or representing words as vectors/arrays of numbers


#### Types
+ Frequency Based
    - CountVectorizer
    - TFIDF
    - Co Occurence
+ Prediction Based
    - Word2Vec
        - CBOW: to predict the probability of a word given a context
        - Skipgram: The aim of skip-gram is to predict the context given a word

```python
# Method 1:
import spacy
nlp = spacy.load('en_core_web_md')
```

```python
mytext = """Under normal conditions, water exists in one of three phases, the solid phase or cold ice, 
the liquid phase water, and the gaseous phase heat steam."""
```

```python
word_tokens = [tok.text for tok in nlp(mytext.lower())]
```

```python
word_tokens
```

```python
tokens = {t.text: Embedding(t.text,t.vector) for t in nlp.pipe(word_tokens)}
```

```python
tokens
```

```python
tokens['water'].vector
```

```python
x_axis = tokens['water']
y_axis = tokens['ice']
for name,t in tokens.items():
    t.plot(x_axis=x_axis,y_axis=y_axis).plot(kind='text',x_axis=x_axis,y_axis=y_axis)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
x_axis = tokens['water']
y_axis = tokens['ice']
for name,t in tokens.items():
    t.plot(x_axis=x_axis,y_axis=y_axis).plot(kind='text',x_axis=x_axis,y_axis=y_axis)
plt.show()
```

### Using Method 2 : SpacyLangauge


```python
from whatlies import EmbeddingSet
from whatlies.language import SpacyLanguage
```

```python
import whatlies.language
dir(whatlies.language)
```

```python
lang = SpacyLanguage('en_core_web_md')
words = ['water','ice','cold','steam','heat','liquid','snow','gaseous']
```

```python
emb = EmbeddingSet(*[lang[w] for w in words ])
```

```python
emb
```

```python
emb['ice'].vector
```

```python
emb.plot_interactive(x_axis=emb['ice'],y_axis=emb['steam'])
```

```python
# Plot Similarity
emb.plot_similarity(metric='cosine')
```
