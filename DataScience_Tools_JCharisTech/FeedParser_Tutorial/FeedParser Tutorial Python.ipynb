```python
#### Convert our RSS Feed to Pandas DF
```

```python
import pandas as pd
```

```python
# Method 1
data = {}
for entry in f.entries:
    data.setdefault('author',[])
    data.setdefault('title',[])
    data.setdefault('summary',[])
    data.setdefault('published_date',[])
    data.setdefault('link',[])
    data['author'].append(entry.author)
    data['title'].append(entry.title)
    data['summary'].append(entry.summary)
    data['published_date'].append(entry.published)
    data['link'].append(entry.link)
    
```

```python
data
```

```python
df = pd.DataFrame(data)
```
